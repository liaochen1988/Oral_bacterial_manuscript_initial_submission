```python
import pandas as pd
from Bio import SeqIO
```

```python
samples = [
    'KP_Exp_8_D0_AVN1',
    'KP_Exp_8_D0_OR_AVN1',
    'KP_Exp_8_D8_AVN1',
    'KP_Exp_8_D0_AVN2',
    'KP_Exp_8_D0_OR_AVN2',
    'KP_Exp_8_D8_AVN2',
    'KP_Exp_8_D0_AVN3',
    'KP_Exp_8_D0_OR_AVN3',
    'KP_Exp_8_D8_AVN3']
```

# Filter count table

```python
df_count = pd.read_csv('qiime2_output/merge_dada2/feature-table/feature-table.from_biom.txt', sep="\t", skiprows=[0], index_col=0)
df_count = df_count[samples].T
df_count = df_count.loc[:, (df_count != 0).any(axis=0)]
df_count.head()
```

```python
df_count.to_csv("filtered_data/count.csv")
```

# Filter taxonomy data

```python
df_tax = pd.read_csv('qiime2_output/merge_dada2/taxonomy.tsv', sep="\t")
df_tax = df_tax[df_tax['Feature ID'].isin(list(df_count.columns))]
df_tax.head()
```

```python
df_tax.to_csv("filtered_data/taxonomy.csv", index=False)
```

# Filter ASV sequences

```python
fasta_sequences = SeqIO.parse(open("qiime2_output/merge_dada2/asv-sequences/dna-sequences.fasta"),'fasta')
with open("filtered_data/dna-sequences.fasta", "w") as out_file:
    for fasta in fasta_sequences:
        name, sequence = fasta.id, str(fasta.seq)
        if name in list(df_tax['Feature ID']):
            out_file.write('>%s\n'%(name))
            out_file.write(sequence)
            out_file.write('\n')
```
