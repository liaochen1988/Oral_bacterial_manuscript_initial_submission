```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from statannot import add_stat_annotation
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
```

# Process meta data

```python
# clean up metadata
# some patients have both IBD and rCDI but their identities are unknown
df_meta = pd.read_csv('qiime2_output/SraRunTable.txt').set_index('Sample Name')
df_meta.index.name = 'SampleID'
df_meta = df_meta[['Assay Type','experimental_timepoint','Organism','Run','Animal_ID','sample_microbial_density_ug_per_mg','condition_at_sampling']]
df_meta = df_meta[df_meta.Organism=='human gut metagenome'].drop(['Organism'], axis=1) # select only human samples
df_meta.columns = ['SequencingAssay','Timepoint','SRA','SubjectID','MicrobialDensity','Condition']
df_meta['SequencingAssay'] = [x.capitalize() for x in df_meta.SequencingAssay]
df_meta['Timepoint'] = ['Day %d'%(x) if str(x)!='nan' and y!='Pouch' else x for x,y in zip(df_meta.Timepoint,df_meta.Condition)]
df_meta['RefDate'] = np.NaN
ref_date = []
for x,y in zip(df_meta.Timepoint, df_meta.Condition):
    if str(x)=='nan' or y=='Pouch':
        ref_date.append(np.NaN)
    else:
        ref_date.append('Fecal microbiota transplantation(%s)'%(y.replace('-FMT','').lower()))
df_meta['RefDate'] = ref_date
df_meta['Treatment'] = np.NaN
df_meta.loc[df_meta.Condition.str.contains('-FMT'), 'Treatment'] = 'Fecal microbiota transplantation-multiple'
df_meta.loc[df_meta.Condition=='Pouch','Treatment'] = 'Ileal pouchâ€“anal anastomosis'
translator = {"Crohn''s Disease":'CD','Healthy':'H','Donor':'H','Post-FMT':'rCDI','Pouch':'UC','Pre-FMT':'rCDI','Ulcerative Colitis':'UC'}
df_meta['Disease'] = [translator[x] for x in df_meta.Condition]
df_meta = df_meta.drop(['Condition'], axis=1)

# add metadata
df_meta2 = pd.read_excel('supp.xlsx', index_col=0)
df_meta2 = df_meta2[['Diagnosis','Antibiotics','Disease Activity']]
df_meta2.columns = ['Disease2','Antibiotic','DetailedDiagnosis']
translator = {"Crohn's Disease":'CD','Healthy':'H','IPAA':'UC','Ulcerative Colitis':'UC'}
df_meta2['Disease2'] = [translator[x] for x in df_meta2.Disease2]
df_meta = pd.merge(df_meta, df_meta2, left_index=True, right_index=True, how='left').reset_index()
new_disease = []
for x,y in zip(df_meta.Disease, df_meta.Disease2):
    if str(x)=='nan' and str(y)=='nan':
        new_disease.append(np.NaN)
    else:
        if str(x)=='nan':
            new_disease.append(y)
        elif str(y)=='nan':
            new_disease.append(x)
        else:
            new_disease.append((',').join(list(set([x,y]))))
df_meta['Disease'] = new_disease
df_meta.loc[df_meta.DetailedDiagnosis==' ','DetailedDiagnosis'] = ''
df_meta.loc[df_meta.DetailedDiagnosis=='Active IBD','DetailedDiagnosis'] = 'Active'
df_meta.loc[df_meta.DetailedDiagnosis=='Inactive IBD','DetailedDiagnosis'] = 'Inactive'
df_meta.loc[df_meta.DetailedDiagnosis=='Healthy','DetailedDiagnosis'] = 'Control'
df_meta = df_meta.set_index('SRA')
df_meta.index.name = 'SRA'
df_meta['BodySite'] = 'Gastrointestinal_tract'
df_meta['BodySubSite'] = 'Stool'
df_meta['Study'] = 'ibd_contijoch2019_elife'
df_meta['QuantitativeAssay'] = 'DNA/mass'
df_meta['PatientGeo'] = 'United States'
df_meta = df_meta.drop(['Disease2'], axis=1)

# taxonomy
df_tax = pd.read_csv('qiime2_output/taxonomy.tsv', sep="\t").rename({'Feature ID':'ASV'}, axis=1)
df_tax = df_tax[df_tax.Taxon.str.contains('d__Bacteria')]
df_tax = df_tax[~(df_tax.Taxon.str.contains('Chloroplast') | df_tax.Taxon.str.contains('Mitochondria'))]
df_tax['LowestTaxonomy'] = [t.split(';')[-1].strip() for t in df_tax.Taxon]

# count
df_count = pd.read_csv('qiime2_output/feature-table/feature-table.from_biom.txt', sep="\t", index_col=0, skiprows=[0])
df_count = df_count.loc[df_tax['ASV']].T
sample_ids_w_suff_reads = list(df_count.sum(axis=1)[df_count.sum(axis=1)>=1000].index)
df_count = df_count.loc[set(df_meta.index).intersection(set(sample_ids_w_suff_reads))]
df_count = df_count.loc[:, (df_count != 0).any(axis=0)]
df_relabun = df_count.div(df_count.sum(axis=1), axis=0)

# oral bacterial fraction
df_blast_HMPv35oral = pd.read_csv("qiime2_output/asv-sequences/blast_HMPv35oral/blast_HMPv35oral_p100.txt", sep="\t", comment="#", header=None)
df_blast_HMPv35oral.columns = ['query_accver', 'subject_accver', 'perc_identity', 'alignment_length', 'mismatches', 'gap_opens', 'qstart', 'qend', 'sstart', 'send', 'evalue', 'bitscore']
df_oral = df_relabun[set(df_relabun.columns).intersection(set(df_blast_HMPv35oral.query_accver))].sum(axis=1).to_frame().rename({0:'OralFrac'}, axis=1)
df_meta_oral = pd.merge(df_meta, df_oral, left_index=True, right_index=True, how='inner')
df_meta_oral['log10_OralFrac'] = np.log10(df_meta_oral['OralFrac'])
df_meta_oral['log10_MicrobialDensity'] = np.log10(df_meta_oral['MicrobialDensity'])
df_meta_oral = df_meta_oral[df_meta_oral.Disease.isin(['UC','CD','H'])] # keep only healthy people, UC and CD patients
df_meta_oral.Antibiotic = df_meta_oral.Antibiotic.fillna('Unknown')
df_meta_oral = df_meta_oral[df_meta_oral.MicrobialDensity>1e-2] # remove samples with too low microbiota density
```

# Fig. 5b

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(7,6))

df2_meta_oral = df_meta_oral[(df_meta_oral.OralFrac>0) & (df_meta_oral.Disease != 'CD')]
_ = sns.scatterplot(x="log10_OralFrac", y="log10_MicrobialDensity", data=df2_meta_oral, hue='Disease', palette='Set2', s=128, edgecolor='k', ax=ax, hue_order=['H','UC'])
_ = sns.regplot(x="log10_OralFrac", y="log10_MicrobialDensity", data=df2_meta_oral, scatter=False, color='k', ax=ax)
slope, intercept, r_value, p_value, std_err = stats.linregress(df2_meta_oral['log10_OralFrac'], df2_meta_oral['log10_MicrobialDensity'])
_ = ax.set_title('Pearson\'s r = %2.2f, P = %2.6f'%(r_value, p_value), fontsize=14)
_ = ax.set_xlim([-4.25, -0.25])
_ = ax.set_ylim([-2.1, -0.1])
_ = ax.set_xticks([-4,-3,-2,-1])
_ = ax.set_xticklabels(['1e-4','1e-3','1e-2','1e-1'], fontsize=12)
_ = ax.set_xlabel('Oral bacterial fraction', fontsize=14)
_ = ax.set_yticks([-2,-1.5,-1,-0.5])
_ = ax.set_yticklabels(['1e-2','1e-1.5','1e-1','1e-0.5'], fontsize=12)
_ = ax.set_ylabel('Microbiome density (mg DNA/g)', fontsize=14)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(3.5,5.5))

df2_meta_oral = df_meta_oral[(df_meta_oral.OralFrac>0)]
df2_meta_oral = df2_meta_oral[df2_meta_oral.Disease != "CD"]

# oral cavity
_ = sns.boxplot(y="log10_OralFrac", x="Disease", data=df2_meta_oral, width=.6, ax=ax[0], order=['H','UC'], palette='Set2', linewidth=1, showfliers=False)
_ = sns.stripplot(y="log10_OralFrac", x="Disease", data=df2_meta_oral, size=2, color=".3", linewidth=0, ax=ax[0], jitter=0.2, palette='Set2', dodge=True, order=['H','UC'])
_ = add_stat_annotation(ax[0], data=df2_meta_oral, y="log10_OralFrac", x="Disease", order=['H','UC'],
                        comparisons_correction=None,
                        box_pairs=[("H","UC")],
                        test='t-test_welch', text_format='star', loc='inside', verbose=1)
_ = ax[0].set_ylim([-4.2,0.2])

# microbial load
_ = sns.boxplot(y="log10_MicrobialDensity", x="Disease", data=df2_meta_oral, width=.6, ax=ax[1], order=['H','UC'], palette='Set2', linewidth=1, showfliers=False)
_ = sns.stripplot(y="log10_MicrobialDensity", x="Disease", data=df2_meta_oral, size=2, color=".3", linewidth=0, ax=ax[1], jitter=0.2, palette='Set2', dodge=True, order=['H','UC'])
_ = add_stat_annotation(ax[1], data=df2_meta_oral, y="log10_MicrobialDensity", x="Disease", order=['H','UC'],
                        comparisons_correction=None,
                        box_pairs=[("H","UC")],
                        test='t-test_welch', text_format='star', loc='inside', verbose=1)
_ = ax[1].set_ylim([-2.1,-0.1])

plt.tight_layout()
```
